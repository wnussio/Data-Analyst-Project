```python
# function is taking word by word from description and compare with KEY WORDS

# in: x as column, require as key word list

def clear_skills(x,require):
    x=str(x)
    x=x.lower()
    signs1=["/","\n","'"]

    signs=['(',')',',',':','.',';','[',']']
    for l in x:
        for s in signs:
            x=x.replace(s,'')
        for s in signs1:
            x=x.replace(s,' ')
    spl=x.split(' ')
    listt=[]

#   i catched problem - some words had few parts ('power bi'), few different spellings ( 'power bi', 'powerBI', etc. ) 
#   loop below is for standarize most important
    
    for word in spl:
        for key in require:
            if word=='report':
                word='reports'
            if key==word:
                if word=='machine': word='machine learning'
                if word=='power' or word=='powerbi': word='power BI'
                if word=='ab': word='ab/mv'
#                 if word==''
                if key=='communication':
                    word='communication skills'
                    
                listt.append(word)
            else:
                continue
    listt = list(dict.fromkeys(listt))
    return listt
```

```python
# using function to catch key words i need for analysis - it is taking a while

df['skills']=df['description'].apply(lambda x: clear_skills(x,key_skills))
df['programs']=df['description'].apply(lambda x: clear_skills(x,key_programs))
df['languages']=df['description'].apply(lambda x: clear_skills(x,comunication))
df['remote']=df['description'].apply(lambda x: clear_skills(x,remote))

df.head()
```

```python
dfs['skills']=dfs['description'].apply(lambda x: clear_skills(x,key_skills))
dfs['programs']=dfs['description'].apply(lambda x: clear_skills(x,key_programs))
dfs['languages']=dfs['description'].apply(lambda x: clear_skills(x,comunication))
dfs['remote']=dfs['description'].apply(lambda x: clear_skills(x,remote))
dfs.head()
```

```python
# because everyone can write in Title whatever he want i decided to standarize it in column 'category'
titles=['Data Analyst','Business Analyst','Master Data Analyst','Junior Data Analyst','DRM Data Analyst','Operations Analyst',
        'Data Governance Analyst','Operations Analyst','Senior Data Analyst','Data Scientist','Web Analyst']

maches=[['junior','data','analyst'],['senior','data','analyst'],['senior','data','engineer'],['junior','data','engineer'],
        ['data','engineer'],['business','analyst'],['data','analyst'],['data','specialist'],['cloud','engineer'],
        ['software','engineer'],['project', 'manager'],['machine', 'learning', 'engineer'],
        ['data','scientist'],['data','analysis'],['credit','management'],['web', 'analyst'],['operations', 'analyst'],
        ['product', 'analyst'],['risk', 'analyst'],['financial', 'analyst' ],['marketing', 'analyst'],['analyst'],['developer']]

# function is similar to clear_skills() but group method is different

def find(title,maches):
    title=title.lower()
    for occ, word in enumerate(maches):   
#         print(occ)
        if all(y in title for y in maches[occ]):
            if maches[occ]==['analyst']: return 'different analyst'
            if maches[occ]==['data', 'analysis']: return 'data analyst'
            ret=str(maches[occ])
            
            signs=[",","'","[","]"]
            for i in signs:
                ret=ret.replace(i,"")
               
            return str(ret)
                                 
    else:
            return 'different'

    
df['category']=df['title'].apply(lambda x: find(x,maches))
dfs['category']=dfs['title'].apply(lambda x: find(x,maches))
```
